```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

x = np.array([1, 2, 3, 4, 5])
y_actual = np.array([2, 4, 5, 4, 5])

m, b = np.polyfit(x, y_actual, 1)
y_predicted = m * x + b

print("Actual y values:    ", y_actual)
print("Predicted y values: ", y_predicted)

mae = mean_absolute_error(y_actual, y_predicted)
mse = mean_squared_error(y_actual, y_predicted)
rmse = np.sqrt(mse)


r_squared = r2_score(y_actual, y_predicted)

n = len(y_actual)
p = 1
adj_r_squared = 1 - (1 - r_squared) * (n - 1) / (n - p - 1)


std_err_estimation = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R^2): {r_squared:.2f}")
print(f"Adjusted R-squared (Adj R^2): {adj_r_squared:.2f}")
print(f"Standard Error of Estimation: {std_err_estimation:.2f}")


plt.scatter(x, y_actual, color='blue', label='Actual')
plt.plot(x, y_predicted, color='red', marker='o', label='Predicted')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Actual vs Predicted Values')
plt.legend()
plt.show()
```
