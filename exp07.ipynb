```python
import numpy as np

def k_medoids_pam(points, initial_medoids):
    def calculate_cost(pts, meds):
        total_cost = 0
        assignments = []
        for p in pts:
            # Manhattan distance to all medoids
            distances = [abs(p - m) for m in meds]
            min_dist = min(distances)
            total_cost += min_dist
            assignments.append(meds[distances.index(min_dist)])
        return total_cost, assignments

    points = np.array(points)
    current_medoids = list(initial_medoids)
    current_cost, _ = calculate_cost(points, current_medoids)

    print(f"Initial Medoids: {current_medoids}, Initial Cost: {current_cost}")

    non_medoids = [p for p in points if p not in current_medoids]
    best_medoids = current_medoids.copy()
    best_cost = current_cost

    # Try every possible swap
    for i in range(len(current_medoids)):
        for o in non_medoids:
            # Create a temporary candidate set of medoids
            candidate_medoids = current_medoids.copy()
            candidate_medoids[i] = o

            new_cost, _ = calculate_cost(points, candidate_medoids)
            print(f"Swap {current_medoids[i]} with {o} -> New Cost: {new_cost}")

            if new_cost < best_cost:
                best_cost = new_cost
                best_medoids = candidate_medoids

    return best_medoids, best_cost

# Inputs
data_points = [2, 4, 5, 10, 12]
medoids_init = [2, 10]

# Execution
final_m, final_c = k_medoids_pam(data_points, medoids_init)

print("\n--- Result ---")
print(f"Final Medoids: {final_m}")
print(f"Final Total Cost: {final_c}")
```
