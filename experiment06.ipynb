```python
import numpy as np

data = np.array([2, 4, 10, 12, 3, 20, 13, 11, 25])

c1= 2
c2 = 20

while True:
  clustor1 = []
  clustor2 = []

  for point in data:
    d1 = abs(point - c1)
    d2 = abs(point - c2)

    if d1 <= d2:
      clustor1.append(point)
    else:
      clustor2.append(point)


    new_c1 = np.mean(clustor1)
    new_c2 = np.mean(clustor2)


  if new_c1 == c1 and new_c2 == c2:
    break

  c1 = new_c1
  c2 = new_c2

print("clustor 1",clustor1)
print("clustor 2",clustor2)

print("final centrid is c1: ",c1)
print("final centrid is c2: ",c2)



```

```python
import numpy as np
from sklearn.cluster import KMeans

data = np.array([
      [2,3],
      [3,4],
      [6,6],
      [7,7]
]
)

kmeans = KMeans(n_clusters=2, init = np.array([[2,3],[6,6]]), n_init=1)
kmeans.fit(data)

print("clusters", kmeans.labels_)
print("centroids", kmeans.cluster_centers_)
```
